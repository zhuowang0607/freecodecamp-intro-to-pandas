## Indexing, Selection and Slicing

Individual columns in the DataFrame can be selected with regular indexing. Each column is represented as a `Series`:

```python
df
```

# Let's start with calling a row with its row name, using `.loc`

```python
# df['Canada'] will NOT work, as this will try to call a column named "Canada".
# Instead, we use .loc for row name and .iloc for row index.
```

```python
df.loc['Canada']
# Well, alough Canada is a valid row name, the returned result is not a row.
# What's worse, it can not even be transposed.
# Because this result is returned as a series.
# This will happen when the returned result only have one row or one column.
# A series always shows as a column.
```

```python
df.loc[['Canada']]
# This returns a df
# A dataframe keeps its original form.
# A row is still a row, a column is still a column.
```

# This time we call a row with its index, using `.iloc`

```python
df.iloc[-1].to_frame().T
# df.iloc[-1] returns a single row. And this single row somehow is reffered as a series but not a dataframe. The problem with this series is that, it will be shown as a fucking column, not a row, and the transpose doesn't wotk on this series either. If I want to display this result as a row, in its original form, I will have to convert this series to a dataframe, then transpose it. Super annoying.
```

```python
df.iloc[[-1]]
```

# Let's call a column this time.

```python
df['Population']
# Here we go again, it is a series now, it can only be displayed as a column, not a row.
```

Note that the `index` of the returned Series is the same as the DataFrame one. And its `name` is the name of the column. If you're working on a notebook and want to see a more DataFrame-like format you can use the `to_frame` method:

```python
df['Population'].to_frame().T
# Again, to_frame, then transpose.
```

```python
df[['Population']]
# Double brackets also do the work.
# A dataframe keeps the data's original form, a column is still column.
```

Multiple columns can also be selected similarly to `numpy` and `Series`:

```python
df[['Population', 'GDP']]
# If calling for multiple rows or columns using a "," , then double brakets are needed.
```

In this case, the result is another `DataFrame`. Slicing works differently, it acts at "row level", and can be counter intuitive:

```python
df.iloc[1:3]
# Now we call the multiple rows with ":", double brakets this time will return errors.
```

Row level selection works better with `loc` and `iloc` **which are recommended** over regular "direct slicing" (`df[:]`).

`loc` selects rows matching the given index:

```python
df.loc[['Italy']]
```

```python
df.loc['France': 'Italy']
```

As a second "argument", you can pass the column(s) you'd like to select:

```python
df.loc['France':'Italy', 'Population']
# A series
```

```python
df.loc['France':'Italy', ['Population']]
# Dataframe
```

```python
df.loc['France':'Italy', 'Population'].to_frame()
```

```python
df.loc['France': 'Italy', ['Population', 'GDP']]
```

```python
df.loc[['Canada', 'Italy'], ['Population', 'GDP']]
```

## Conditional selection (boolean arrays)

We saw conditional selection applied to `Series` and it'll work in the same way for `DataFrame`s. After all, a `DataFrame` is a collection of `Series`:

```python
df
```

```python
df['Population'] > 70
# series
```

```python
df[['Population']] > 70
# dataframe
```

```python
df.loc[df['Population'] > 70]
```

The boolean matching is done at Index level, so you can filter by any row, as long as it contains the right indexes. Column selection still works as expected:

```python
df.loc[df['Population'] > 70, ['Population']]
```

```python
df.loc[df['Population'] > 70, ['Population', 'GDP']]
```

## Dropping stuff

Opposed to the concept of selection, we have "dropping". Instead of pointing out which values you'd like to _select_ you could point which ones you'd like to `drop`:

# This `drop` won't operate in place, meaning the original 'df' is not affected.

```python
df.drop('Canada')
```

```python
df.drop(['Canada', 'Japan'])
# Dropping Rows
```

```python
df
# See, Japan and Canada are still here.
```

```python
df.drop(columns=['Population', 'HDI'])
# Dropping columns
```

```python
df.drop(['Italy', 'Canada'], axis=0)
# axis=0 indicats rows
```

```python
df.drop(['Population', 'HDI'], axis=1)
# axis=1 indicats columns
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
df.drop(['Population', 'HDI'], axis='columns')
```

```python
df.drop(['Canada', 'Germany'], axis='rows')
```

```python
myindex = list(range(0,3)) + [5]
myindex
```

```python
df.index[myindex]
```

```python
df.drop(df.index[myindex], axis='rows')
```

```python
df.drop(df.index[[1,3]], axis='rows')
```

```python
df.drop(df.columns[[1,3]], axis='columns')
```

```python
df
```

```python
df.columns[[1,3]]
```

```python
# Drop rows from myindex, and columns 'Surface Area' and 'Language'
df.drop(index=df.index[myindex], columns=df.columns[[1,3]])
```

## Creating columns from other columns

Altering a DataFrame often involves combining different columns into another. For example, in our Countries analysis, we could try to calculate the "GDP per capita", which is just, `GDP / Population`.

```python
df[['Population', 'GDP']]
```

The regular pandas way of expressing that, is just dividing each series:

```python
df['GDP'] / df['Population']
```

The result of that operation is just another series that you can add to the original `DataFrame`:

```python
df['GDP Per Capita'] = df['GDP'] / df['Population']
```

```python
df
```
